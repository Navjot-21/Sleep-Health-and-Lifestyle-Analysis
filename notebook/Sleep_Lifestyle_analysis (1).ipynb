To investigate how sleep duration and sleep quality are related to stress levels, physical activity, BMI, blood pressure, occupation, and sleep disorders.

# Sleep Characteristics
## Understanding Sleep Patterns

```python
data['Sleep Category'] = pd.cut(
    data['Sleep Duration'],
    bins=[0,6,8,10],
    labels=['Poor Sleep','Normal Sleep','Good Sleep']
)
data['Sleep Category'].value_counts()
```

```python
#Distribution of Sleep Duration
fig,ax=plt.subplots(1,2,figsize=(12,5))

# Histogram
ax[0].hist(data['Sleep Duration'],bins=10)
ax[0].set_title('Distribution of Sleep Duration')
ax[0].set_xlabel('Sleep Duration')
ax[0].set_ylabel('Frequency')

# KDE Plot
sns.kdeplot(data['Sleep Duration'],fill=True,ax=ax[1])
ax[1].set_title('Density of Sleep Duration')
ax[1].set_xlabel('Sleep Duration')

plt.tight_layout()
plt.show()
```

```python
#Sleep Categories Distribution
fig,ax=plt.subplots(1,2,figsize=(12,5))

# Countplot
sns.countplot(x='Sleep Category',data=data,ax=ax[0])
ax[0].set_title('Sleep Categories')

# Pie Chart
data['Sleep Category'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    ax=ax[1]
)
ax[1].set_ylabel('')

plt.tight_layout()
plt.show()
```

```python
#sleep duration vs Quality of Sleep
plt.figure(figsize=(12,5))
sns.regplot(
    x='Sleep Duration',
    y='Quality of Sleep',
    data=data,)
plt.show()
```


### Observation

Most individuals sleep between 6 and 8 hours. Sleep quality generally improves with increasing sleep duration.

These findings emphasize the importance of maintaining regular physical activity and an active lifestyle for achieving and sustaining a healthy body composition.

Overall, the results suggest that physical activity, daily movement, and BMI are strongly interconnected. Lower levels of activity and fewer daily steps are associated with higher BMI, highlighting the role of an active lifestyle in promoting better health and well-being.

## Blood Pressure Analysis

```python
fig,ax = plt.subplots(1,2,figsize=(12,5))

# Systolic
sns.histplot(
    data['Systolic'],
    bins=10,
    kde=True,
    ax=ax[0]
)
ax[0].set_title('Distribution of Systolic Pressure')

# Diastolic
sns.histplot(
    data['Diastolic'],
    bins=10,
    kde=True,
    ax=ax[1]
)
ax[1].set_title('Distribution of Diastolic Pressure')

plt.tight_layout()
plt.show()
```

```python
fig,ax = plt.subplots(1,2,figsize=(12,5))

sns.regplot(
    x='Sleep Duration',
    y='Systolic',
    data=data,
    ax=ax[0]
)
ax[0].set_title('Sleep Duration vs Systolic BP')

sns.regplot(
    x='Sleep Duration',
    y='Diastolic',
    data=data,
    ax=ax[1]
)
ax[1].set_title('Sleep Duration vs Diastolic BP')

plt.tight_layout()
plt.show()
```

```python
fig,ax = plt.subplots(1,2,figsize=(12,5))

sns.regplot(
    x='Stress Level',
    y='Systolic',
    data=data,
    ax=ax[0]
)
ax[0].set_title('Stress Level vs Systolic BP')

sns.regplot(
    x='Stress Level',
    y='Diastolic',
    data=data,
    ax=ax[1]
)
ax[1].set_title('Stress Level vs Diastolic BP')

plt.tight_layout()
plt.show()
```

```python
fig,ax = plt.subplots(1,2,figsize=(12,5))

sns.barplot(
    x='BMI Category',
    y='Systolic',
    data=data,
    ax=ax[0]
)
ax[0].set_title('Average Systolic BP')
ax[0].tick_params(axis='x',rotation=45)

sns.barplot(
    x='BMI Category',
    y='Diastolic',
    data=data,
    ax=ax[1]
)
ax[1].set_title('Average Diastolic BP')
ax[1].tick_params(axis='x',rotation=45)

plt.tight_layout()
plt.show()
```

### Observation

The blood pressure analysis reveals meaningful relationships between sleep duration, stress levels, and BMI categories. The distribution plots indicate that most individuals have systolic blood pressure concentrated around 125–135 mmHg and diastolic blood pressure around 80–90 mmHg, suggesting that the majority of individuals fall within a relatively narrow blood pressure range.

The regression analysis shows a slight negative relationship between sleep duration and both systolic and diastolic blood pressure. Individuals with longer sleep durations tend to exhibit marginally lower blood pressure levels, indicating that adequate sleep may contribute to improved cardiovascular health.

Conversely, stress level demonstrates a positive association with blood pressure. As stress levels increase, both systolic and diastolic blood pressure show a tendency to rise, highlighting the impact of psychological stress on cardiovascular functioning.

BMI category analysis further supports this relationship. Individuals with normal BMI exhibit the lowest average systolic and diastolic blood pressure, whereas obese individuals display the highest values. Overweight individuals fall between these two groups. These findings suggest that unhealthy body weight is associated with increased cardiovascular risk.

Overall, the results indicate that sleep duration, stress levels, and BMI are interconnected factors influencing blood pressure. Adequate sleep and healthy body composition, combined with effective stress management, appear to contribute to better cardiovascular health and lower blood pressure levels.

Cardiovascular indicators also show noticeable differences among the sleep disorder categories. Individuals with insomnia and sleep apnea tend to exhibit relatively higher systolic and diastolic blood pressure values than individuals without any disorder, indicating an increased risk of cardiovascular complications.

### Observation

Occupation appears to have a significant influence on sleep patterns and stress levels. The analysis reveals noticeable variations in average sleep duration, stress levels, and sleep quality across different professions. Occupations associated with higher stress levels generally tend to exhibit lower sleep duration and poorer sleep quality, whereas professions with relatively lower stress levels demonstrate healthier sleep patterns.

These differences suggest that work-related demands, lifestyle, and stress associated with different occupations may indirectly affect sleep health and overall well-being. Individuals experiencing greater occupational stress are more likely to have reduced sleep duration, which can subsequently impact sleep quality and other health indicators.

```python
num_cols = [
'Age',
'Sleep Duration',
'Quality of Sleep',
'Physical Activity Level',
'Stress Level',
'Heart Rate',
'Daily Steps',
'Systolic',
'Diastolic'
]

plt.figure(figsize=(10,8))

sns.heatmap(
    data[num_cols].corr(),
    annot=True,
    cmap='coolwarm'
)

plt.show()
```

### Observation

The correlation heatmap provides an overall view of the relationships among the numerical variables and reveals several significant patterns.

Sleep Duration and Quality of Sleep exhibit a strong positive correlation (0.88), indicating that individuals who sleep longer generally experience better sleep quality. 

In contrast, Stress Level shows a strong negative correlation with both Sleep Duration (-0.81) and Quality of Sleep (-0.90), suggesting that increased stress is associated with shorter and poorer-quality sleep.

Heart Rate demonstrates a moderate positive correlation with Stress Level (0.67) and a negative correlation with Sleep Duration (-0.52) and Quality of Sleep (-0.66). These relationships indicate that stress and inadequate sleep may adversely affect cardiovascular health.

Physical Activity Level and Daily Steps show a strong positive correlation (0.77), confirming that individuals who are more physically active tend to accumulate more daily steps.
Physical activity also exhibits weak positive relationships with sleep-related variables, suggesting that an active lifestyle may contribute to better sleep.

Systolic and Diastolic Blood Pressure are highly positively correlated (0.97), indicating that both measures tend to increase together. 

Moreover, blood pressure shows moderate positive correlations with age and weak positive correlations with stress and heart rate, highlighting the combined influence of aging and lifestyle factors on cardiovascular health.

Overall, the heatmap confirms that sleep duration, sleep quality, stress level, physical activity, heart rate, and blood pressure are closely interconnected, supporting the findings obtained throughout the analysis.




# Overall Conclusion

This study explored the relationship between sleep patterns and various lifestyle and health indicators, including stress levels, physical activity, BMI, blood pressure, occupation, age, gender, and sleep disorders. The analysis demonstrates that sleep is closely interconnected with multiple aspects of physical and mental well-being rather than being an isolated phenomenon.

The results reveal that adequate sleep is associated with better sleep quality, lower stress levels, healthier heart rate and blood pressure, and improved overall lifestyle. Individuals with higher physical activity levels and healthier BMI categories generally exhibit more favorable sleep characteristics, whereas elevated stress levels are strongly associated with poor sleep quality and increased cardiovascular risk.

Furthermore, demographic and occupational factors influence sleep behavior indirectly through their impact on stress and lifestyle. Individuals suffering from sleep disorders tend to experience poorer sleep quality, higher stress levels, and elevated blood pressure, emphasizing the importance of maintaining healthy sleep habits.

Overall, the findings highlight that good sleep, regular physical activity, stress management, and healthy lifestyle choices are essential components of long-term physical and mental well-being. Sleep should therefore be considered a fundamental pillar of health, as its effects extend beyond rest and influence nearly every aspect of an individual's lifestyle and quality of life.
